```python
import math
import os

def get_dir_size(path='.', names=[], sizes=[], indent=0):
    total = 0
    with os.scandir(path) as it:
        for entry in it:
            if entry.is_file():
                total += entry.stat().st_size
            elif entry.is_dir():
                sub_total, subnames, subsizes = get_dir_size(entry.path, [], [], indent+1)
                total += sub_total
                names.extend(subnames)
                sizes.extend(subsizes)
    order = int(math.log(total+1, 1000))
    total_str = f'{total/1000**order:.2f}'[:4].strip('.')
    names.insert(0, f'{"  "*indent}{path.split("/")[-1]}' )
    sizes.insert(0, f'{total_str} {" KMGT"[order]}B')
    return total, names, sizes

def print_folder_sizes(path='.', min_size='K'):
    '''set min_size={'M', 'G', or 'T'} to only show bigger folders'''
    _, names, sizes = get_dir_size(path)
    max_len = len(max(names, key=len))
    for name, size in zip(names, sizes):
        if ' KMGT'.index(size[-2]) >= ' KMGT'.index(min_size):
            print(name, ' '*(max_len-len(name)), size.rjust(7))

print_folder_sizes()
```
